### Concatenate Data Frames

```python
import pandas as pd
india_weather = pd.DataFrame({
    'city':['mumbai','bangalore','delhi'],
    'temperature':[32,27,45],
    'humidity':[80,60,79]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'city':['nyc','chicago','palo alto'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
us_weather
```

```python
# concatenate two dataframes
df = pd.concat([india_weather , us_weather])
df
```

```python
# continuous index
df = pd.concat([india_weather , us_weather], ignore_index=True)
df
```

```python
df = pd.concat([india_weather,us_weather],axis=1)
df
```

### Merge DataFrames

```python
temperature_df = pd.DataFrame({
    'city':['mumbai','delhi','bangalore','hyderabad'],
    'temperature':[35,40,32,37]
})

temperature_df
```

```python
humdity_df = pd.DataFrame({
    'city':['delhi','mumbai','bangalore'],
    'humidity':[68,65,75]
})
humdity_df
```

```python
# merge two dataframes without explicitly mention index
df = pd.merge(temperature_df,humdity_df,on='city')
df
```

```python
# Outer-join
df = pd.merge(temperature_df,humdity_df,on='city',how='outer')
df
```
